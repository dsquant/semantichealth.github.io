Get (simpler) terms in parentheses

```python
import re
```

```python
def get_condition(condition_tuple):
    condition = condition_tuple[0]
    match = re.search('\((.*?)\)', condition)
    if match:
        return match.group(0).strip('()')
    else:
        return condition    
```

```python
conditions_simple = map(get_condition, conditions_array)
```

```python
conditions_simple[:10]
```

```python
conditions_list = list(set(conditions_simple))
```

```python
conditions_list[:20]
```

```python
json.dump(conditions_list, open("conditions.json", 'w'))
```
